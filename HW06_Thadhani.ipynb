```r
readStates <- function(x) # Running the Function after it is called
  
{
  
  #Removing Unneded rows by row selector
  
  Statesfun <- Statesfun[-1:-8,]
  
  Statesfun <- Statesfun[-52:-58,]
  
  rownames(Statesfun) <- NULL
  
  dim(Statesfun)
  
  #Removing Unneded columns by column selector
  
  Statesfun <- Statesfun[,-6:-10]
  
  dim(Statesfun)
  
  # Renaming column names using colnames function 
  
  colnames(Statesfun)
  
  
  colnames(Statesfun)[1] <- "stateName"
  colnames(Statesfun)[2] <- "Census"
  colnames(Statesfun)[3] <- "Estimated"
  colnames(Statesfun)[4] <- "Pop2010"
  colnames(Statesfun)[5] <- "Pop2011"
  
  colnames(Statesfun)
  
  dim(Statesfun)
  
  # Using gsub command to remove dot from stateName, comma from other columns 
  
  Statesfun$stateName <- gsub("\\.","",Statesfun$stateName)
  
  Statesfun$Census <- gsub(",","",Statesfun$Census)
  
  Statesfun$Estimated <- gsub(",","",Statesfun$Estimated)
  
  Statesfun$Pop2010 <- gsub(",","",Statesfun$Pop2010)
  
  Statesfun$Pop2011 <- gsub(",","",Statesfun$Pop2011)
  
  # Using as.numeric command
  
  Statesfun$Census <- as.numeric(gsub(" ","",Statesfun$Census))
  
  Statesfun$Estimated <- as.numeric(gsub(" ","",Statesfun$Estimated))
  
  Statesfun$Pop2010 <- as.numeric(gsub(" ","",Statesfun$Pop2010))
  
  Statesfun$Pop2011 <- as.numeric(gsub(" ","",Statesfun$Pop2011))
  
  return(Statesfun)
  
}
```
